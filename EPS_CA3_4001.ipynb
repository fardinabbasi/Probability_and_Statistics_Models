# Q1: Bertrand Paradox

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

## Part1

```python
counter= 0
point1=np.random.uniform(0,2*math.pi,1000)
point2=np.random.uniform(0,2*math.pi,1000)

s_angle1=np.sin(point1)
c_angle1=np.cos(point1)
s_angle2=np.sin(point2)
c_angle2=np.cos(point2)
```

```python
figure, axes = plt.subplots()
for i in range (999):
    x = [c_angle1[i],c_angle2[i]]
    y = [s_angle1[i],s_angle2[i]]
    r = math.sqrt((c_angle1[i]-c_angle2[i])**2+(s_angle1[i]-s_angle2[i])**2)
    if (r < 1.732):
        plt.plot(x, y, 'k')
    else:
        plt.plot(x, y, 'y')
        counter=counter+ 1
Drawing_uncolored_circle = plt.Circle( (0, 0 ),1)
axes.set_aspect( 1 )
axes.add_artist( Drawing_uncolored_circle )
plt.title('Circle(part1)')
plt.show()
print('Probability(part1)=',counter/ 1000)
```

## Part 2

```python
counter= 0
teta= np.random.uniform(0,2*math.pi,1000)
P = np.random.uniform(0,1,1000)

Angle1 = teta - np.arccos(P)
Angle2 = teta + np.arccos(P)

s_angle1= np.sin(Angle1)
c_angle1= np.cos(Angle1)
s_angle2= np.sin(Angle2)
c_angle2= np.cos(Angle2)

Q=2*np.sqrt(1 - np.power(P,2))
```

```python
figure, axes = plt.subplots()
for i in range (999):
    x = [c_angle1[i],c_angle2[i]]
    y = [s_angle1[i],s_angle2[i]]
    r = math.sqrt((c_angle1[i] -c_angle2[i])**2 + (s_angle1[i]-s_angle2[i])**2)
    if (Q[i] < 1.732):
        plt.plot(x, y, 'k')
    else:
        counter = counter + 1
        plt.plot(x, y, 'y')
Drawing_uncolored_circle = plt.Circle( (0, 0 ),1)
axes.set_aspect( 1 )
axes.add_artist( Drawing_uncolored_circle )
plt.title('Circle(part2)')
plt.show()
print('Probability(part2)=',counter/ 1000)
```

# Q2: Estimation of Euler Number with Uniform Distribution

```python
import numpy as np
from matplotlib import pyplot as plt
import random
```

```python
n=10000
Y=0
ouler=[None]*n
counter=[0]*n

for i in range(n):
    Y=0
    while(Y<1):
        X=random.uniform(0,1)
        Y=Y+X
        counter[i]=counter[i]+1
    ouler[i]=(sum(counter))/(i+1)
```

```python
x=range(1,n+1,1)
y=np.array(ouler)

plt.figure()
plt.plot(x,y,'b',label='Euler')
plt.title('Plotter')
plt.xlabel('n')
plt.ylabel('Euler')
plt.axhline(y=2.71828, linestyle=":", color="red")
plt.show()
```

# Q3: Banach's Match-Box Problem

```python
import random
from matplotlib import pyplot as plt
```

```python
N =60
maches= [0] * N
x = [0] * N
box= [0] * 2
trials= 10000

for i in range (trials-1):
    box[0] = N
    box[1] = N
    while((box[0] > 0) and (box[1] > 0)):
        r = random.randint(0,1)
        box[r] = box[r] - 1
    if (box[0] == 0):
        maches[box[1]] =maches[box[1]] + 1
    else:
        maches[box[0]] =maches[box[0]] + 1
for i in range (N):
    x[i] = i
plt.bar(x,maches,color ='g')
plt.ylabel('maches')
plt.show()
```

 # Q4: Integral Calculation

```python
import numpy as np
from matplotlib import pyplot as plt
import random
import math
import abc
```

```python
n=4000000

#calculate integral 0:1 x^3 dx
x=np.random.uniform(0,1,n)
f=np.power(x,3)
F=((1-0)/(n-1))*sum(f)
eror=abs(F-0.25)
print(f'integral 0:1 x^3={F},eror={eror}')#wolfram answer=0.25
```

```python
#calculate integral 0:pi sin(x) dx
x=np.random.uniform(0,math.pi,n)
f=np.sin(x)
F=((math.pi-0)/(n-1))*sum(f)
eror=abs(F-2)
print(f'integral 0:pi sin(x)={F},eror={eror}')#wolfram answer=2
```

```python
#calculate integral 0:4 1/(1+x^2)dx
x=np.random.uniform(0,4,n)
f=np.power((1+np.power(x,2)),-1)
F=((4-0)/(n-1))*sum(f)
eror=abs(F-1.3258)
print(f'integral 0:4 1/(1+x^2)={F},eror={eror}')#wolfram answer=1.3258
```

```python
#calculate integral 0:1 sqrt(1-x^4)dxdx
x=np.random.uniform(0,1,n)
f=np.sqrt(1-np.power(x,4))
F=((1-0)/(n-1))*sum(f)
print('integral 0:1 sqrt(1-x^4)dx is:',F)#wolfram answer=0.874019
```

```python
#calculate integral 0:1 e^(x^2)dx
x=np.random.uniform(0,1,n)
f=np.exp(np.power(x,2))
F=((1-0)/(n-1))*sum(f)
print('integral 0:1 e^(x^2)dx is:',F)#wolfram answer=1.46265
```
